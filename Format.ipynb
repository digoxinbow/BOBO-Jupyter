# 格式插入字串 .format()

https://www.w3schools.com/python/ref_string_format.asp

The format() method formats the specified value(s) and insert them inside the string's placeholder.

The placeholder is defined using curly brackets: {}. 

```python
print( 'poo-poo {}' .format('FACE') )
```

```python
print( 'poo-poo {} {} {}' .format( 'FACE' , 'FINGER' , 'FOOT' ) )
```

```python
# 花括弧編號索引(numbered indexes)

print( 'poo-poo {2} {1} {0}' .format( 'FACE' , 'FINGER' , 'FOOT' ) )
print( 'poo-poo {1} {1} {1}' .format( 'FACE' , 'FINGER' , 'FOOT' ) )
```

```python
# 花括弧命名索引(named indexes)

print( 'poo-poo {o} {i} {a}' .format( a = 'FACE' , i = 'FINGER' , o = 'FOOT' ) )
print( 'poo-poo {i} {i} {i}' .format( a = 'FACE' , i = 'FINGER' , o = 'FOOT' ) )
```

# 浮點數格式

{變數位置:[總寬度].[小數位數]f}

f是float代表浮點數



```python
my_BMI = 50 / 1.6 / 1.6
my_BMI
```

```python
# {變數位置:[總寬度].[小數位數]f}
print("Ffion's BMI is {B:.2f}".format(B=my_BMI))
print("Ffion's BMI is {B:.1f}".format(B=my_BMI))
print("Ffion's BMI is {B:.0f}".format(B=my_BMI))
```

```python
# 整個字串的總寬度，多加數值會得到空格
print("Ffion's BMI is {B:6.2f}".format(B=my_BMI))
print("Ffion's BMI is {B:12.2f}".format(B=my_BMI))
```
