### Why doesn't 0.1+0.2-0.3 equal 0.0 ?

因為計算機架構中是以二進位浮點數來表示小數的噢！

我們輸入的十進位浮點數只能由實際儲存在計算機中的二進位浮點數近似。

```python
0.1+0.2-0.3
```

Python是一種動態型別的程式語言，這意味著變數不需要提前宣告**數據類型**。

這是Python提供的靈活性，但也使得編碼和調試過程中錯誤的風險增加。雙面刃！

BOBO另外查的round()函數！

可將數字四捨五入到指定的小數位數

round( 欲四捨五入的數字 , 指定小數點後位數 )

```python
my_height = 160
my_weight = 50
my_BMI = round ( ( my_weight / ( my_height / 100 ) ** 2 ) , 2 )
```

# String 字串

字串（string），是由零個或多個字元組成的有順序的序列（Ordered Sequences）。

它是程式語言中表示文字的資料型別。

\n 換行

\t tab

len( ) 返回序列（如字符串、列表、元组）的長度

```python
print( "Ffion & \nEthan" )
print( "Ffion & \tEthan" )
len ( "Ffion & Ethan" )
```
